```python
import pandas as pd

prem_table = pd.read_html('https://www.bbc.co.uk/sport/football/premier-league/table')
Premier_table = prem_table[0]

print(len(prem_table))
print(type(prem_table))

Premier_table.head(6)
```

```python
df = Premier_table.drop(['Unnamed: 1'], axis=1)
df.head(6)
```

```python
df.tail(3)
```

```python
df.drop(df.tail(1).index, inplace=True)
df.shape
```

```python
df.tail(3)
```

```python
df['Goal Ratio'] = round(df['F'].astype(int)/df['A'].astype(int), 1)
```

```python
df.head(3)
```

```python
df.rename(columns={'Unnamed: 0':'Positon'}, inplace=True)
```

```python
df = df[['Positon','Team','P','W','D','L','F','A','GD','Pts','Goal Ratio']]
```
